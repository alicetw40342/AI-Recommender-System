## ğŸ“– Task 2 - Transform Text into AI Embeddings

In this task, I transform text data into numerical vector embeddings using the **all-MiniLM-L6-v2** model from `sentence-transformers`. I process at least 50 plot summaries from my dataset (`movie_plots.csv`) and store the resulting embeddings using NumPy arrays. These embeddings will later be used for similarity-based recommendations.

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd

# ä½¿ç”¨ä¸Šå‚³çš„æª”æ¡ˆåç¨±
file_name = list(uploaded.keys())[0]

# è¼‰å…¥è³‡æ–™
df = pd.read_csv(file_name)

# é¡¯ç¤ºè³‡æ–™ç­†æ•¸èˆ‡å‰å¹¾ç­†
print("ç¸½å…±æœ‰å¹¾ç­†è³‡æ–™ï¼š", len(df))
df.head()
```

```python
# å‡è¨­æ¬„ä½åç‚º "Plot"ï¼Œå–å‰ 50 ç­†æ–‡å­—å…§å®¹
texts = df['Plot'].dropna().head(50).tolist()

print("å¯¦éš›å–å‡ºçš„æ–‡å­—æ•¸é‡ï¼š", len(texts))
print("å‰ 1 ç­†å…§å®¹ï¼š\n", texts[0])
```

```python
from sentence_transformers import SentenceTransformer

# è¼‰å…¥é è¨“ç·´æ¨¡å‹
model = SentenceTransformer('all-MiniLM-L6-v2')

# å°‡æ–‡å­—è½‰æ›ç‚º embeddings å‘é‡
embeddings = model.encode(texts, show_progress_bar=True)

print("åµŒå…¥å‘é‡çš„å½¢ç‹€ï¼š", embeddings.shape)
```

```python
import numpy as np
import pickle

# å„²å­˜ç‚º .npy
np.save("movie_plot_embeddings.npy", embeddings)

# å„²å­˜ç‚º .pklï¼ˆå¯é¸ï¼‰
with open("movie_plot_embeddings.pkl", "wb") as f:
    pickle.dump(embeddings, f)

print("å‘é‡å„²å­˜å®Œæˆï¼")
```
